```
import pandas as pd

# Load the Excel file
file_path = '/bin/Datafile_2.xlsx'
df = pd.read_excel(file_path)

# Display the first few rows of the DataFrame to check column names
print("Original DataFrame Columns:")
print(df.columns)

# Get the actual column names from the DataFrame
actual_columns = df.columns.tolist()

# Define a mapping from expected column names to actual column names
column_mapping = {
    'Name': 'Name',  # Replace 'Name' with the actual column name for 'Name'
    'Address': 'Address',  # Replace 'Address' with the actual column name for 'Address'
    'Zip': 'Zip',  # Replace 'Zip' with the actual column name for 'Zip'
    'Location': 'Location'  # Replace 'Location' with the actual column name for 'Location'
}

# Clean the data using the actual column names
for expected_column, actual_column in column_mapping.items():
    if actual_column in actual_columns:
        # Strip whitespace from string columns only if they exist in the DataFrame
        if df[actual_column].dtype == 'object':
            df[actual_column] = df[actual_column].str.strip()
        # Convert 'Zip' to string and strip whitespace only if it exists in the DataFrame
        if expected_column == 'Zip':
            df[actual_column] = df[actual_column].astype(str).str.strip()

# Extract structured data from the Address column
# This regex assumes Address format is "Street, City, StateAbbreviation"
# Adjust if your format is different
if 'Address' in actual_columns:
    df[['Street', 'City', 'State']] = df['Address'].str.extract(r'^(.*),\s*(.*),\s*([A-Z]{2})')

# Display the cleaned and structured DataFrame
print("\nCleaned and Structured DataFrame:")
print(df)
```
